```python
# Import libraries and dependencies
import pandas as pd
from pathlib import Path

# Import the hvPlot library
import hvplot.pandas
import holoviews

# Path leading to the CSV file required to analyze paychecks
tmDB_comedy_csv = Path("Resources/filtered_tmDB_data_2010-24_comedy.csv")
tmDB_horror_csv = Path("Resources/filtered_tmDB_data_2010-24_horror.csv")

tmDB_comedy_df = pd.read_csv(tmDB_comedy_csv)
tmDB_comedy_df = tmDB_comedy_df.sort_values(by='release_date', ascending=True)

tmDB_horror_df = pd.read_csv(tmDB_horror_csv)
tmDB_horror_df = tmDB_horror_df.sort_values(by='release_date', ascending=True)
```

```python
tmDB_comedy_plot = tmDB_comedy_df.hvplot(
    x='release_date',
    y='popularity',
    color='blue',  # Choose a color for comedy movies
    title="Popularity Over Time (Comedy)",
    xlabel="Release Date",
    ylabel="Popularity"
)

tmDB_horror_plot = tmDB_horror_df.hvplot(
    x='release_date',
    y='popularity',
    color='red',  # Choose a color for horror movies
    title="Popularity Over Time (Horror)",
    xlabel="Release Date",
    ylabel="Popularity"
)
```
